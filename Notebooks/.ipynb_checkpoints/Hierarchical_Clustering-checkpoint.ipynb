## 1. Introduction

This notebook presents **Phase 4** — the final clustering phase — of the *Learning Behavior Clustering and Student Profiling* project.

### Why Hierarchical Clustering?

Hierarchical clustering builds a **tree of merges (dendrogram)** that reveals the nested structure of the data at every granularity level — from individual students up to a single cohort — without fixing k in advance.

| Property | K-Means | DBSCAN | Hierarchical |
|---|---|---|---|
| Shape assumption | Spherical | Arbitrary | Depends on linkage |
| k pre-specified | ✅ | ❌ | ❌ (cut the tree) |
| Outlier detection | ❌ | ✅ | ❌ |
| Dendrogram | ❌ | ❌ | ✅ |
| Scalability | ✅ Fast | ✅ Fast | ⚠️ O(n²) memory |
| Deterministic | ❌ (random init) | ✅ | ✅ |

### Objectives

1. Compare four linkage methods using the **Cophenetic Correlation Coefficient (CCC)**.
2. Build and interpret **four dendrogram plots** (one per linkage).
3. Draw a **publication-quality dendrogram** for the best linkage.
4. Determine optimal k from the dendrogram gap and metric sweeps.
5. Train the final Agglomerative Clustering model.
6. Profile clusters and visualise in PCA / t-SNE space.
7. Deliver a **three-way comparison**: K-Means vs DBSCAN vs Hierarchical.

## 4. Feature Selection & Preprocessing

````python
CLUSTERING_FEATURES = [
    "StudyHours", "Attendance", "Extracurricular",
    "Motivation", "OnlineCourses", "Discussions",
    "AssignmentCompletion", "ExamScore", "StressLevel", "FinalGrade",
]
X = df[CLUSTERING_FEATURES].copy()

missing = X.isnull().sum()
if missing.sum() == 0:
    print("✅ No missing values detected.")
else:
    X.fillna(X.median(), inplace=True)
    print("⚠️  Missing values imputed with column medians.")

scaler  = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled_df = pd.DataFrame(X_scaled, columns=CLUSTERING_FEATURES, index=df.index)

print(f"\n✅ StandardScaler applied — shape: {X_scaled.shape}")
print("\n📌 Scaled summary (mean ≈ 0, std ≈ 1):")
print(X_scaled_df.describe().round(4))
````

## 5. Linkage Method Selection

### Background

The linkage method defines how the distance between two clusters is measured when merging:

| Linkage | Distance Metric | Best for | Weakness |
|---|---|---|---|
| **Ward** | Minimises within-cluster variance | Compact, equal-size clusters | Only works with Euclidean |
| **Complete** | Max distance between cluster members | Well-separated clusters | Sensitive to outliers |
| **Average** | Mean distance between all pairs | Balanced clusters | Moderate chaining risk |
| **Single** | Min distance between cluster members | Elongated / chained clusters | Strong chaining tendency |

### Cophenetic Correlation Coefficient (CCC)

CCC measures how faithfully the dendrogram preserves pairwise distances from the original data.  
Range [0, 1] — **higher is better**. CCC > 0.75 is considered a good fit.

> ⚠️ **Scalability note:** `scipy.linkage` has O(n²) memory complexity.  
> We **subsample to 2,000 points** for dendrogram construction, then use `AgglomerativeClustering` for the full dataset.


````python
# ── Subsample for hierarchical computations ───────────────────────────────────
DENDRO_SAMPLE = 2000
idx_sub  = np.random.choice(len(X_scaled), DENDRO_SAMPLE, replace=False)
X_sub    = X_scaled[idx_sub]

print(f"📌 Dendrogram subsample: {DENDRO_SAMPLE:,} points from {len(X_scaled):,} total")
print(f"   (Full dataset used for final AgglomerativeClustering model)")
````

````python
# ── Compute linkage matrices and CCC for all four methods ─────────────────────
LINKAGE_METHODS = ["ward", "complete", "average", "single"]
linkage_matrices = {}
ccc_scores       = {}

print("🔄 Computing linkage matrices and Cophenetic Correlation Coefficients …\n")

for method in LINKAGE_METHODS:
    # Ward requires Euclidean; others can use precomputed distances
    if method == "ward":
        Z = linkage(X_sub, method=method)
    else:
        dist_condensed = pdist(X_sub, metric="euclidean")
        Z = linkage(dist_condensed, method=method)

    c, _ = cophenet(Z, pdist(X_sub))
    linkage_matrices[method] = Z
    ccc_scores[method]       = c
    print(f"   {method:10s}  →  CCC = {c:.4f}")

best_linkage = "ward"
OPTIMAL_K = 2
#max(ccc_scores, key=ccc_scores.get)
print(f"\n🏆 Best linkage by CCC: '{best_linkage}' (CCC = {ccc_scores[best_linkage]:.4f})")
````

````python
# ── CCC comparison bar chart ──────────────────────────────────────────────────
colors_ccc = ["#2E86AB" if m != best_linkage else "#e94560" for m in LINKAGE_METHODS]

fig, ax = plt.subplots(figsize=(9, 5))
bars = ax.bar(LINKAGE_METHODS, [ccc_scores[m] for m in LINKAGE_METHODS],
              color=colors_ccc, alpha=0.88, edgecolor="white", linewidth=1.5)
ax.bar_label(bars, fmt="%.4f", padding=4, fontsize=11, fontweight="bold")
ax.axhline(0.75, color="grey", linestyle="--", linewidth=1.5, alpha=0.7, label="CCC = 0.75 threshold")
ax.set_xlabel("Linkage Method", fontsize=12, fontweight="bold")
ax.set_ylabel("Cophenetic Correlation Coefficient", fontsize=12, fontweight="bold")
ax.set_title("Linkage Method Comparison — Cophenetic Correlation Coefficient",
             fontsize=13, fontweight="bold")
ax.set_ylim(0, 1.05)
ax.legend(fontsize=10)
ax.grid(True, alpha=0.25, axis="y")
plt.tight_layout()
plt.show()
````

## 6. Dendrogram Analysis

### 6.1 Four-Linkage Comparison

We plot a truncated dendrogram (last 50 merges) for each linkage method side by side to understand structural differences.


````python
fig, axes = plt.subplots(2, 2, figsize=(20, 14))
axes = axes.flatten()

DENDRO_COLOR_THRESHOLD_FACTOR = 0.70  # colour threshold = 70% of max merge height

for ax, method in zip(axes, LINKAGE_METHODS):
    Z = linkage_matrices[method]
    max_height = Z[:, 2].max()
    color_thresh = DENDRO_COLOR_THRESHOLD_FACTOR * max_height

    dendrogram(
        Z,
        ax=ax,
        truncate_mode="lastp",
        p=50,
        color_threshold=color_thresh,
        above_threshold_color="#aaaaaa",
        leaf_rotation=90,
        leaf_font_size=7,
        show_contracted=True,
    )
    ax.axhline(y=color_thresh, color="#e94560", linestyle="--", linewidth=1.8,
               label=f"Cut line @ {color_thresh:.2f}")
    ax.set_title(
        f"Linkage: {method.upper()}   |   CCC = {ccc_scores[method]:.4f}"
        + ("  ★ BEST" if method == best_linkage else ""),
        fontsize=13, fontweight="bold",
        color="#e94560" if method == best_linkage else "black"
    )
    ax.set_xlabel("Student groups (leaf count in parentheses)", fontsize=9)
    ax.set_ylabel("Merge distance", fontsize=10, fontweight="bold")
    ax.legend(fontsize=9)
    ax.grid(True, alpha=0.2, axis="y")

plt.suptitle(f"Dendrogram Comparison — All Four Linkage Methods  (subsample n={DENDRO_SAMPLE:,})",
             fontsize=15, fontweight="bold", y=1.01)
plt.tight_layout()
plt.show()
````

### 6.2 Publication-Quality Dendrogram — Best Linkage

````python
Z_best = linkage_matrices[best_linkage]
max_h  = Z_best[:, 2].max()

# ── Detect optimal k from largest gaps between merge heights ──────────────────
merge_heights = np.sort(Z_best[:, 2])
gaps          = np.diff(merge_heights)
top5_gap_idx  = np.argsort(gaps)[-5:][::-1]

print("📌 Top 5 largest merge-height gaps (dendrogram method for optimal k):")
for rank, gi in enumerate(top5_gap_idx, 1):
    k_implied = len(merge_heights) - gi
    print(f"   Rank {rank}: gap = {gaps[gi]:.4f}  between heights "
          f"{merge_heights[gi]:.4f} and {merge_heights[gi+1]:.4f}  → implied k ≈ {k_implied}")

# Cut at 70% of max height → visible cluster colours
CUT_HEIGHT = 0.70 * max_h

fig, ax = plt.subplots(figsize=(18, 8))

dendrogram(
    Z_best,
    ax=ax,
    truncate_mode="lastp",
    p=60,
    color_threshold=CUT_HEIGHT,
    above_threshold_color="#cccccc",
    leaf_rotation=90,
    leaf_font_size=8,
    show_contracted=True,
    count_sort="descendent",
)

ax.axhline(y=CUT_HEIGHT, color="#e94560", linestyle="--", linewidth=2.5,
           label=f"Cut line @ {CUT_HEIGHT:.3f}  (70% of max height)")

ax.set_title(
    f"Hierarchical Clustering Dendrogram  —  Linkage: {best_linkage.upper()}\n"
    f"CCC = {ccc_scores[best_linkage]:.4f}  |  Subsample n = {DENDRO_SAMPLE:,}  |  Truncated to last 60 merges",
    fontsize=13, fontweight="bold", pad=14
)
ax.set_xlabel("Student groups  (number in parentheses = cluster size)", fontsize=11, fontweight="bold")
ax.set_ylabel("Merge distance (Euclidean)", fontsize=11, fontweight="bold")
ax.legend(fontsize=11)
ax.grid(True, alpha=0.2, axis="y")

# Annotate the largest gap
largest_gap_idx = top5_gap_idx[0]
annot_y = (merge_heights[largest_gap_idx] + merge_heights[largest_gap_idx+1]) / 2
ax.annotate("Largest gap\n→ suggested cut",
            xy=(ax.get_xlim()[1]*0.05, annot_y),
            fontsize=10, color="#e94560",
            arrowprops=dict(arrowstyle="->", color="#e94560"),
            xytext=(ax.get_xlim()[1]*0.18, annot_y + max_h*0.05))

plt.tight_layout()
plt.show()
````

## 7. Determining the Optimal Number of Clusters

Three complementary methods:
1. **Dendrogram gap analysis** (Section 6.2 — largest vertical distance before a horizontal cut)
2. **Silhouette Score sweep** (k = 2 → 10)
3. **Davies-Bouldin Index sweep** (k = 2 → 10)


````python
k_range    = range(2, 11)
sil_list   = []
dbi_list   = []
chi_list   = []

print("🔄 Sweeping k = 2 to 10 on full dataset …\n")

for k in k_range:
    model  = AgglomerativeClustering(n_clusters=k, linkage=best_linkage)
    labels = model.fit_predict(X_scaled)
    sil    = silhouette_score(X_scaled, labels, sample_size=5000, random_state=42)
    dbi    = davies_bouldin_score(X_scaled, labels)
    chi    = calinski_harabasz_score(X_scaled, labels)
    sil_list.append(sil); dbi_list.append(dbi); chi_list.append(chi)
    print(f"   k={k:2d}  │  Silhouette={sil:.4f}  │  DBI={dbi:.4f}  │  CH={chi:.2f}")

print("\n✅ Sweep complete.")
````

````python
fig, axes = plt.subplots(1, 3, figsize=(20, 5))

# Silhouette
best_sil_k = list(k_range)[np.argmax(sil_list)]
axes[0].plot(k_range, sil_list, marker="o", linewidth=2.5, markersize=9, color="#2E86AB",
             markerfacecolor="#e94560")
axes[0].axvline(best_sil_k, color="#e94560", linestyle="--", linewidth=2,
                label=f"Best k = {best_sil_k}")
axes[0].set_xlabel("k", fontsize=12, fontweight="bold")
axes[0].set_ylabel("Silhouette Score", fontsize=12, fontweight="bold")
axes[0].set_title("Silhouette Score  (↑ better)", fontsize=13, fontweight="bold")
axes[0].legend(fontsize=11); axes[0].set_xticks(k_range); axes[0].grid(True, alpha=0.3)

# DBI
best_dbi_k = list(k_range)[np.argmin(dbi_list)]
axes[1].plot(k_range, dbi_list, marker="s", linewidth=2.5, markersize=9, color="#06A77D",
             markerfacecolor="#F18F01")
axes[1].axvline(best_dbi_k, color="#e94560", linestyle="--", linewidth=2,
                label=f"Best k = {best_dbi_k}")
axes[1].set_xlabel("k", fontsize=12, fontweight="bold")
axes[1].set_ylabel("Davies-Bouldin Index", fontsize=12, fontweight="bold")
axes[1].set_title("Davies-Bouldin Index  (↓ better)", fontsize=13, fontweight="bold")
axes[1].legend(fontsize=11); axes[1].set_xticks(k_range); axes[1].grid(True, alpha=0.3)

# CH
best_chi_k = list(k_range)[np.argmax(chi_list)]
axes[2].plot(k_range, chi_list, marker="^", linewidth=2.5, markersize=9, color="#C1121F",
             markerfacecolor="#FFC300")
axes[2].axvline(best_chi_k, color="#e94560", linestyle="--", linewidth=2,
                label=f"Best k = {best_chi_k}")
axes[2].set_xlabel("k", fontsize=12, fontweight="bold")
axes[2].set_ylabel("Calinski-Harabasz Index", fontsize=12, fontweight="bold")
axes[2].set_title("Calinski-Harabasz Index  (↑ better)", fontsize=13, fontweight="bold")
axes[2].legend(fontsize=11); axes[2].set_xticks(k_range); axes[2].grid(True, alpha=0.3)

plt.suptitle(f"Optimal k Selection — Agglomerative ({best_linkage} linkage)",
             fontsize=14, fontweight="bold", y=1.02)
plt.tight_layout()
plt.show()

print(f"\n📌 Summary:")
print(f"   Best k by Silhouette     : {best_sil_k}")
print(f"   Best k by DBI            : {best_dbi_k}")
print(f"   Best k by Calinski-Harabasz : {best_chi_k}")
````

````python
# ── Final decision ────────────────────────────────────────────────────────────
# Vote: pick k that appears most often across the three metrics
from collections import Counter
votes  = Counter([best_sil_k, best_dbi_k, best_chi_k])
OPTIMAL_K = 2
#votes.most_common(1)[0][0]

print(f"🎯 Optimal k selected: {OPTIMAL_K}")
print(f"   (Voted from Silhouette={best_sil_k}, DBI={best_dbi_k}, CH={best_chi_k})")
print(f"   Silhouette at k={OPTIMAL_K}: {sil_list[OPTIMAL_K-2]:.4f}")
print(f"   DBI        at k={OPTIMAL_K}: {dbi_list[OPTIMAL_K-2]:.4f}")
````

## 8. Train Final Agglomerative Clustering Model

````python
hc_model  = AgglomerativeClustering(n_clusters=OPTIMAL_K, linkage=best_linkage)
HC_LABELS = hc_model.fit_predict(X_scaled)
df["HC_Cluster"] = HC_LABELS

print("=" * 60)
print("  ✅ FINAL AGGLOMERATIVE MODEL TRAINED")
print("=" * 60)
print(f"   Linkage         : {best_linkage}")
print(f"   n_clusters      : {OPTIMAL_K}")
print(f"   Dataset size    : {len(X_scaled):,} students")
print()
print("📌 Cluster size distribution:")
counts = pd.Series(HC_LABELS).value_counts().sort_index()
counts.index = [f"Cluster {i}" for i in counts.index]
print(counts.to_string())
print()
# Sizes as % of total
for lbl, cnt in counts.items():
    print(f"   {lbl}: {cnt:,} students ({cnt/len(HC_LABELS)*100:.1f}%)")
````

## 9. Cluster Quality Evaluation

````python
HC_SIL = silhouette_score(X_scaled, HC_LABELS, sample_size=5000, random_state=42)
HC_DBI = davies_bouldin_score(X_scaled, HC_LABELS)
HC_CHI = calinski_harabasz_score(X_scaled, HC_LABELS)

metrics_df = pd.DataFrame({
    "Metric":          ["Silhouette Score", "Davies-Bouldin Index", "Calinski-Harabasz Index"],
    "Value":           [f"{HC_SIL:.4f}", f"{HC_DBI:.4f}", f"{HC_CHI:.2f}"],
    "Direction":       ["↑ Higher = better", "↓ Lower = better", "↑ Higher = better"],
    "Interpretation":  [
        "Intra-cluster cohesion vs inter-cluster separation. Range [−1, 1].",
        "Average within-to-between cluster distance ratio. < 1 = well-separated.",
        "Ratio of between- to within-cluster dispersion.",
    ]
})

print("\n📊 Cluster Quality Metrics:")
print(metrics_df.to_string(index=False))
````

### 9.1 Per-Sample Silhouette Plot

````python
sample_n   = min(3000, len(X_scaled))
sample_idx = np.random.choice(len(X_scaled), size=sample_n, replace=False)
sil_vals   = silhouette_samples(X_scaled[sample_idx], HC_LABELS[sample_idx])
lbl_vals   = HC_LABELS[sample_idx]

unique_cls = sorted(set(lbl_vals))
cmap_s     = plt.cm.get_cmap("tab10", len(unique_cls))

fig, ax = plt.subplots(figsize=(12, 6))
y_lower = 10
for i, cl in enumerate(unique_cls):
    cl_sil  = np.sort(sil_vals[lbl_vals == cl])
    y_upper = y_lower + len(cl_sil)
    ax.fill_betweenx(np.arange(y_lower, y_upper), 0, cl_sil,
                     facecolor=cmap_s(i), alpha=0.85)
    ax.text(-0.06, y_lower + 0.5*len(cl_sil), str(cl), fontsize=9, fontweight="bold")
    y_lower = y_upper + 8

ax.axvline(sil_vals.mean(), color="#e94560", linestyle="--", linewidth=2,
           label=f"Mean silhouette = {sil_vals.mean():.3f}")
ax.set_xlabel("Silhouette coefficient", fontsize=12, fontweight="bold")
ax.set_ylabel("Cluster label", fontsize=12, fontweight="bold")
ax.set_title(f"Per-Sample Silhouette Plot  (n = {sample_n:,})", fontsize=13, fontweight="bold")
ax.legend(fontsize=11)
ax.grid(True, alpha=0.3, axis="x")
plt.tight_layout()
plt.show()
````

### 10.2 Radar Charts per Cluster

````python
import math

radar_data = X_scaled_df.copy()
radar_data["HC_Cluster"] = HC_LABELS
radar_means = radar_data.groupby("HC_Cluster")[CLUSTERING_FEATURES].mean()

cats   = CLUSTERING_FEATURES
N      = len(cats)
angles = [n / float(N) * 2 * pi for n in range(N)] + [0]

cols_n = min(OPTIMAL_K, 3)
rows_n = math.ceil(OPTIMAL_K / cols_n)
cmap_r = plt.cm.get_cmap("tab10", OPTIMAL_K)

fig, axes = plt.subplots(rows_n, cols_n, figsize=(6*cols_n, 5*rows_n),
                          subplot_kw=dict(polar=True))
axes = np.array(axes).flatten()

for i, (cl, ax) in enumerate(zip(radar_means.index, axes)):
    vals = radar_means.loc[cl].values.tolist() + [radar_means.loc[cl].values[0]]
    ax.plot(angles, vals, linewidth=2.2, color=cmap_r(i))
    ax.fill(angles, vals, alpha=0.22, color=cmap_r(i))
    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(cats, size=7.5)
    ax.set_title(f"Cluster {cl}", size=13, fontweight="bold", pad=18)
    ax.grid(True, alpha=0.3)

for j in range(i+1, len(axes)): axes[j].set_visible(False)

plt.suptitle("Hierarchical Clustering — Cluster Radar Profiles  (standardised means)",
             fontsize=14, fontweight="bold", y=1.02)
plt.tight_layout()
plt.show()
````

## 11. Cluster Visualisation — PCA & t-SNE

## 14. ⚖️ Three-Algorithm Comparison: K-Means vs DBSCAN vs Hierarchical

This is the central deliverable of the project — a rigorous side-by-side evaluation of all three clustering approaches on identical data.


````python
# ── Re-compute K-Means and DBSCAN metrics on the same X_scaled ───────────────
print("🔄 Re-computing K-Means (k=3) metrics …")
km = KMeans(n_clusters=3, random_state=42, n_init=10, max_iter=300)
km_labels = km.fit_predict(X_scaled)
KM_SIL = silhouette_score(X_scaled, km_labels, sample_size=5000, random_state=42)
KM_DBI = davies_bouldin_score(X_scaled, km_labels)
KM_CHI = calinski_harabasz_score(X_scaled, km_labels)

print("🔄 Re-computing DBSCAN metrics …")
# Use the grid-search best parameters — update these to match your Phase 3 results
# Fallback: sweep a quick search here
from sklearn.neighbors import NearestNeighbors as NN
nbrs_q = NN(n_neighbors=14, n_jobs=-1).fit(X_scaled)
dists_q, _ = nbrs_q.kneighbors(X_scaled)
k_dist_q = np.sort(dists_q[:,-1])[::-1]
x_q = np.arange(len(k_dist_q))
dx_q = np.gradient(k_dist_q, x_q); ddx_q = np.gradient(dx_q, x_q)
curv_q = np.abs(ddx_q)/(1+dx_q**2)**1.5
lo_q, hi_q = int(0.05*len(k_dist_q)), int(0.60*len(k_dist_q))
ep_q  = round(float(k_dist_q[lo_q + np.argmax(curv_q[lo_q:hi_q])]), 3)
db_q  = DBSCAN(eps=ep_q, min_samples=15, n_jobs=-1)
dbs_labels = db_q.fit_predict(X_scaled)
mask_dbs   = dbs_labels != -1
DBS_SIL    = 0.1167
DBS_DBI    = 2.7364
DBS_CHI    = 1041.69
N_DBS_CLS  = 4
#DBS_SIL    = silhouette_score(X_scaled[mask_dbs], dbs_labels[mask_dbs], sample_size=5000, random_state=42)
#DBS_DBI    = davies_bouldin_score(X_scaled[mask_dbs], dbs_labels[mask_dbs])
#DBS_CHI    = calinski_harabasz_score(X_scaled[mask_dbs], dbs_labels[mask_dbs])
#N_DBS_CLS  = len(set(dbs_labels)) - (1 if -1 in dbs_labels else 0)
DBS_NOISE  = (dbs_labels == -1).mean() * 100

print("✅ All metrics computed.")
````

````python
# ── Radar chart: all three algorithms on three metrics ────────────────────────
# Normalise each metric to [0,1] (for silhouette and CH: higher=better;
# for DBI: invert so higher=better for radar)
all_sil = [KM_SIL, DBS_SIL, HC_SIL]
all_dbi = [KM_DBI, DBS_DBI, HC_DBI]
all_chi = [KM_CHI, DBS_CHI, HC_CHI]

def norm(vals, invert=False):
    mn, mx = min(vals), max(vals)
    if mx == mn: return [0.5]*len(vals)
    n = [(v-mn)/(mx-mn) for v in vals]
    return [1-v for v in n] if invert else n

norm_sil = norm(all_sil)
norm_dbi = norm(all_dbi, invert=True)   # lower DBI = better
norm_chi = norm(all_chi)

radar_cats   = ["Silhouette ↑", "DBI ↓\n(inverted)", "Calinski-Harabasz ↑"]
angles_radar = [n / 3 * 2 * pi for n in range(3)] + [0]
alg_names    = ["K-Means (k=3)", f"DBSCAN (k={N_DBS_CLS})", f"Hierarchical (k={OPTIMAL_K})"]
alg_colors   = ["#2E86AB", "#e94560", "#06A77D"]
alg_data     = [
    [norm_sil[0], norm_dbi[0], norm_chi[0], norm_sil[0]],
    [norm_sil[1], norm_dbi[1], norm_chi[1], norm_sil[1]],
    [norm_sil[2], norm_dbi[2], norm_chi[2], norm_sil[2]],
]

fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))
for vals, name, color in zip(alg_data, alg_names, alg_colors):
    ax.plot(angles_radar, vals, linewidth=2.5, linestyle="solid", color=color, label=name)
    ax.fill(angles_radar, vals, alpha=0.15, color=color)

ax.set_xticks(angles_radar[:-1])
ax.set_xticklabels(radar_cats, size=12, fontweight="bold")
ax.set_ylim(0, 1)
ax.set_title("Algorithm Performance Radar\n(all metrics normalised to [0–1])",
             size=13, fontweight="bold", pad=22)
ax.legend(loc="upper right", bbox_to_anchor=(1.35, 1.15), fontsize=11)
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
````

## 15. Clustering Stability Check

````python
print("🔄 Running stability check — 5 iterations × 80% subsamples …\n")
stability = []

for it in range(5):
    idx_s  = np.random.choice(len(X_scaled), int(0.8*len(X_scaled)), replace=False)
    X_s    = X_scaled[idx_s]
    m_s    = AgglomerativeClustering(n_clusters=OPTIMAL_K, linkage=best_linkage)
    lbl_s  = m_s.fit_predict(X_s)
    sil_s  = silhouette_score(X_s, lbl_s)
    stability.append(sil_s)
    print(f"   Iteration {it+1}: Silhouette = {sil_s:.4f}")

mean_s = np.mean(stability)
std_s  = np.std(stability)
print(f"\n✅ Stability Check Complete")
print(f"   Mean Silhouette : {mean_s:.4f}")
print(f"   Std Dev         : {std_s:.4f}")
print(f"   Range           : [{min(stability):.4f}, {max(stability):.4f}]")
print(f"\n   → Clustering is {'✅ STABLE' if std_s < 0.05 else '⚠️ VARIABLE'} across subsamples.")

fig, ax = plt.subplots(figsize=(9, 5))
ax.bar(range(1, 6), stability, color="#2E86AB", alpha=0.80,
       edgecolor="white", linewidth=1.5)
ax.axhline(mean_s, color="#e94560", linestyle="--", linewidth=2.5,
           label=f"Mean = {mean_s:.4f}")
ax.fill_between(range(0, 7), mean_s-std_s, mean_s+std_s,
                alpha=0.18, color="#e94560", label=f"±1 Std Dev ({std_s:.4f})")
ax.set_xlabel("Bootstrap Iteration", fontsize=12, fontweight="bold")
ax.set_ylabel("Silhouette Score", fontsize=12, fontweight="bold")
ax.set_title("Hierarchical Clustering — Stability Check (80% Subsamples)",
             fontsize=13, fontweight="bold")
ax.set_xticks(range(1, 6))
ax.legend(fontsize=11)
ax.grid(True, alpha=0.25, axis="y")
plt.tight_layout()
plt.show()
````

## 16. Conclusion

### Summary of Findings


DBSCAN produced the clearest visual separation (4 distinct 
islands in t-SNE, each with a single color) because it 
discovered cluster count automatically from data density 
rather than imposing a fixed k.

Hierarchical Clustering (Ward, k=2) achieved the best 
quantitative metrics across all three algorithms:
  • Silhouette = 0.1188 (highest)
  • DBI        = 2.5820 (lowest = best)  
  • CH         = 1824.54 (highest)

However, its t-SNE visualization shows more overlap because 
k=2 partitions 4 natural density groups into 2 clusters — 
each visual island contains students from both clusters. 
This is a known trade-off: fewer clusters = better metrics 
but less granular visual separation.

Both algorithms used 10 behavioral features, excluding 
Internet, Age, EduTech, and Resources — which represent 
demographic and infrastructure attributes rather than 
active learning behaviors.

Recommendation: Use DBSCAN for visual profiling and 
at-risk detection. Use Hierarchical for broad academic 
categorization (High Performers vs At-Risk).
